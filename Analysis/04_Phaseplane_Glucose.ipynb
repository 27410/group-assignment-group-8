Phenotype phase plane analysis

```python
from cobra.flux_analysis.phenotype_phase_plane import production_envelope
%run 01_Model_CFE.ipynb
```


Defining the glucose content in the YEPD media:


```python
#Changing flux of succrose
Medium=iMM904.medium
Medium['EX_glc__D_e']=20
iMM904.reactions.EX_glc__D_e.bounds=-20,1000
```

Constructing the phenotype phase plane plot for the biomass objective:

```python

PPP_biomass = production_envelope(iMM904,
                    reactions=[iMM904.reactions.EX_glc__D_e],
                    objective=iMM904.reactions.BIOMASS_SC5_notrace)
PPP_biomass
```

```python
PPP_biomass.plot(x='EX_glc__D_e', y='flux_maximum', ylabel="Biomass")
```

```python
PPP_AA = production_envelope(iMM904,
                    reactions=[iMM904.reactions.EX_glc__D_e],
                    objective=iMM904.reactions.ALDH1)
PPP_AA
```
