```python
from cobra.flux_analysis.phenotype_phase_plane import production_envelope
%run Model_CFE.ipynb
```

Constructing the phenotype phase plane plot for the biomass objective:

```python
#Changning flux of glucose
Medium=iMM904.medium
Medium['EX_sucr_e']=20
iMM904.reactions.EX_sucr_e.bounds=-20,1000

PPP_biomass = production_envelope(iMM904,
                    reactions=[iMM904.reactions.EX_sucr_e],
                    objective=iMM904.reactions.BIOMASS_SC5_notrace)
PPP_biomass
```

```python
PPP_biomass.plot(x='EX_sucr_e', y='flux_maximum', ylabel="Biomass")
```

Here is seen that there ia a linear relationship with glucose and the amount of biomass produced
